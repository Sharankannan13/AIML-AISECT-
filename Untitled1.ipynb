```python
import pandas as pd
import numpy as np
dataset=pd.read_csv('/content/Project 1 - Weather Dataset.csv')
```

```python
dataset
```

```python
dataset=dataset.drop(["Date/Time"],axis = 1)
```

```python
dataset
```

```python
dataset=dataset.drop(["Dew Point Temp_C"],axis = 1)
```

```python
dataset
```

```python
def map_weather_keywords(w):
    w = w.lower()
    if 'clear' in w:
        return 'Clear'
    elif 'cloudy' in w:
        return 'Cloudy'
    elif 'rain' in w or 'drizzle' in w or 'shower' in w:
        return 'Rain'
    elif 'snow' in w or 'pellet' in w:
        return 'Snow'
    elif 'fog' in w or 'haze' in w:
        return 'Fog/Mist/Haze'
    elif 'thunder' in w:
        return 'Thunderstorm'
    elif 'freezing' in w or 'ice' in w:
        return 'Ice/Freezing'
    else:
        return 'Other'

dataset['Weather_Category'] = dataset['Weather'].apply(map_weather_keywords)
display(dataset['Weather_Category'].value_counts())
```

```python
dataset.columns
```

```python
dataset=dataset.drop(["Weather"],axis = 1)
```

```python
X = dataset[['Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Visibility_km', 'Press_kPa']]
Y = dataset['Weather_Category']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.33, random_state=42)
```

```python
X
```

```python
Y
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
X_train
```

```python
from sklearn.metrics import classification_report
```

```python
cls=classification_report(y_test, y_pred)
```
